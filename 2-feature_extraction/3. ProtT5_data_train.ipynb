#### Proses Generate Negative

```python
import pandas as pd
import random

def generate_negative_interactions(positive_interactions, num_negative_interactions):
    """
    Membuat pasangan interaksi negatif dari dataset positif.
    
    Parameters:
    - positive_interactions (DataFrame): DataFrame dengan pasangan positif (pep_seq, Uniprot_id).
    - num_negative_interactions (int): Jumlah pasangan negatif yang diinginkan.
    
    Returns:
    - DataFrame: DataFrame berisi pasangan negatif (pep_seq, Uniprot_id).
    """
    unique_peptides = positive_interactions['pep_seq'].unique()
    unique_proteins = positive_interactions['Uniprot_id'].unique()

    positive_pairs = set(zip(positive_interactions['pep_seq'], positive_interactions['Uniprot_id']))

    negative_interactions = set()

    while len(negative_interactions) < num_negative_interactions:
        # Pilih secara acak peptida dan protein
        peptide = random.choice(unique_peptides)
        protein = random.choice(unique_proteins)

        if (peptide, protein) not in positive_pairs and (peptide, protein) not in negative_interactions:
            negative_interactions.add((peptide, protein))

    return pd.DataFrame(list(negative_interactions), columns=['pep_seq', 'Uniprot_id'])
```

```python
# Generating negative interactions 1:1
num_negative_interactions = 3470
negative_interactions = generate_negative_interactions(data, num_negative_interactions)
```
