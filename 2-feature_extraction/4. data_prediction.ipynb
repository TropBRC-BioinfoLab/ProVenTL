### Data Protein Cancer

```python
prot_entries= 'D:/CODING/Code_new/Data Testing/Prot_Cancer_ProtT5.csv'
prot_entries = pd.read_csv(prot_entries)
prot_entries
```

```python
pep_prot_pairs = pd.DataFrame(list(product(pep_entries.iterrows(), prot_entries.iterrows())),
                              columns=["pep_row", "prot_row"])

pep_prot_pairs = pep_prot_pairs.sample(frac=1, random_state=42).reset_index(drop=True)

# 3. Extract and combine features
def combine_features(row):
    pep_row = row['pep_row'][1] 
    prot_row = row['prot_row'][1]

    combined_data = {
        'pdb_chain': pep_row['pdb_chain'], 
        'seq': pep_row['Seq'],             
        'Uniprot_id': prot_row['Uniprot_id'],  
        'prot_seq': prot_row['prot_seq'],      #
        **{f'pep_embed_{i}': pep_row[f'pep_embed_{i}'] for i in range(1024)},  
        **{f'prot_embed_{i}': prot_row[f'prot_embed_{i}'] for i in range(1024)},  
    }
    return combined_data

combined_features = pd.DataFrame(pep_prot_pairs.apply(combine_features, axis=1).tolist())

combined_features['label'] = "" 
```

```python
testing_data_path = 'D:/CODING/Code_new/Data Testing/data_testing_PepPI.csv'
combined_features.to_csv(testing_data_path, index=False)

print(f"Dataset testing acak dengan metadata telah disimpan di: {testing_data_path}")
```
