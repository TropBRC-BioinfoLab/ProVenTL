```python
import pandas as pd
from joblib import load
from tensorflow.keras.models import load_model
```

```python
testing_file_path = '/home/g6523023/data_testing_PepPI/data_testing_PepPI.csv'
data_testing = pd.read_csv(testing_file_path)
data_testing
```

```python
X_test = data_testing.drop(columns=['label', 'pdb_chain', 'seq', 'Uniprot_id', 'prot_seq'])
```

```python
X_test
```

```python
model_path = "/home/g6523023/Percobaan Final/DL/hasil_dnn/Retraining_new/best_model_auc_0.8748.h5"
loaded_model = load_model(model_path)
```

```python
# Prediksi probabilitas
predictions_probabilities = loaded_model.predict(X_test)
```

```python
threshold = 0.9
predictions_classes = (predictions_probabilities >= threshold).astype(int)
```

```python
results_df = data_testing[['pdb_chain', 'seq', 'Uniprot_id', 'prot_seq']].copy()
results_df['predict_probability'] = predictions_probabilities.flatten()
results_df['predict'] = predictions_classes.flatten()
```

```python
import pandas as pd
hasil_prediksi_1_path = '/home/g6523023/Percobaan Final/Prediksi/class1_1_prediction.csv'
hasil_prediksi_1 = pd.read_csv(hasil_prediksi_1_path)
hasil_prediksi_1
```

```python
mean_val = hasil_prediksi_1['predict_probability'].mean()
std_val = hasil_prediksi_1['predict_probability'].std()
```

```python
hasil_prediksi_1['Z_Score'] = (hasil_prediksi_1['predict_probability'] - mean_val) / std_val
```

```python
from scipy.stats import norm
hasil_prediksi_1['p_value'] = 1 - norm.cdf(hasil_prediksi_1['Z_Score'])
```

```python
interaksi_signifikan = hasil_prediksi_1[
    (hasil_prediksi_1['predict_probability'] >= 0.95) &
    (hasil_prediksi_1['p_value'] < 0.05)
]
```
